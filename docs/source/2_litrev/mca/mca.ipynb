```python
import prince
import pandas as pd
import matplotlib.pyplot as plt
```

### Importation of the data

```python
dataset = pd.read_csv("models.csv")
dataset = dataset.set_index("Model")
dataset = dataset[['Model type', 'Geographical scope', 'Solution concept',
       'Solution horizon', 'Solution method']]
```

```python
dataset
```

```python
dataset = dataset.dropna()
```

```python
mca = prince.MCA(

    n_components=6, 
    n_iter=3, 
    copy=True, 
    check_input=True, 
    engine='sklearn',
    random_state=42

)

mca = mca.fit(dataset)
```

```python
one_hot = pd.get_dummies(dataset)

mca_no_one_hot = prince.MCA(one_hot=False)
mca_no_one_hot = mca_no_one_hot.fit(one_hot)
```

```python
mca.eigenvalues_summary
```

```python
mca.row_coordinates(dataset).head()
```

```python
ax = mca.plot(
    dataset, 
    show_row_labels=True, 
    show_row_markers=False, 
    show_column_labels=False,
    show_column_markers=False
    )

ax.save('pca.png', ppi=200, scale_factor=2)
ax
```
