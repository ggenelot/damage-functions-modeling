```python
import pandas as pd

# Supposons que vous ayez déjà chargé votre DataFrame 'shapes'
# shapes = pd.read_csv('votre_fichier.csv')  # Ou toute autre méthode de chargement

# Filtrer et trier les colonnes
df_filtered = shapes[['Name', 'Equation number', 'Model', 'Latex fragment', 'Input variable', 'Equation form']]

# Trier par 'Equation form'
df_filtered = df_filtered.sort_values(by='Equation form')
df_filtered = df_filtered.dropna(subset=['Equation form'])

# Générer la table LaTeX avec des séparations par 'Equation form'
def generate_latex_table(df):
    latex_code = []
    
    # Début de la table LaTeX
    latex_code.append(r'\begin{table}[ht]')
    latex_code.append(r'\centering')
    latex_code.append(r'\begin{longtable}{|l|l|l|l|l|}')
    latex_code.append(r'\hline')
    latex_code.append(r' Model & Equation number & Name & Latex Fragment & Input Variables \\')
    latex_code.append(r'\hline')

    current_equation_form = None

    # Remplir la table avec les données
    for index, row in df.iterrows():
        # Si l'Equation form change, insérer une ligne vide avec le nom de la forme centré
        if row['Equation form'] != current_equation_form:
            current_equation_form = row['Equation form']
            latex_code.append(r'\multicolumn{5}{c}{\textbf{' + current_equation_form + r'}} \\')
            latex_code.append(r'\hline')
        
        name = row['Name']
        eq_number = row['Equation number']
        model = row['Model']
        latex_fragment = f"${row['Latex fragment']}$"  # Mettre le fragment dans un environnement mathématique
        input_variable = row['Input variable']
        
        # Ajouter chaque ligne de données au tableau
        latex_code.append(f' {model} &  {eq_number} & {name}  & {latex_fragment} & {input_variable} \\\\')
        latex_code.append(r'\hline')
    
    # Fin de la table LaTeX
    latex_code.append(r'\end{tabular}')
    latex_code.append(r'\caption{Table des fonctions de dommage}')
    latex_code.append(r'\label{tab:damage_functions}')
    latex_code.append(r'\end{table}')
    
    # Retourner le code LaTeX
    return '\n'.join(latex_code)

# Générer la table LaTeX
latex_table = generate_latex_table(df_filtered)

# Écrire dans un fichier .tex
with open('../../../../Memoire/results/table_damage_functions.tex', 'w') as f:
    f.write(latex_table)

# Afficher le code LaTeX généré
print(latex_table)
```

```python
import pandas as pd

# Supposons que vous ayez déjà chargé votre DataFrame 'shapes'
# shapes = pd.read_csv('votre_fichier.csv')  # Ou toute autre méthode de chargement

# Filtrer et trier les colonnes
df_filtered = shapes[['Model', 'Equation number', 'Name', 'Latex fragment', 'Input variable', 'Equation form']]

# Trier par 'Equation form'
df_filtered = df_filtered.sort_values(by='Equation form')
df_filtered = df_filtered.dropna(subset=['Equation form'])
df_filtered['Latex fragment'] = df_filtered['Latex fragment'].fillna('')
df_filtered['Input variable'] = df_filtered['Input variable'].fillna('Unknown')


# Fonction pour vérifier la validité d'un fragment LaTeX

def is_valid_latex(fragment):

    try:

        # Vérifier la compilation du fragment LaTeX

        from pylatexenc.latexwalker import LatexWalker

        lw = LatexWalker(fragment)

        lw.get_latex_nodes()
        print(f'Equation compiled : {fragment}')

        return True  # Si la compilation réussit, c'est valide

    except Exception:

        print(f'Invalid equation : {fragment}')

        return False  # Si une erreur survient, le fragment n'est pas valide




# Générer la table LaTeX avec longtable
def generate_latex_table(df):
    latex_code = []
    
    # Début de la table LaTeX avec longtable
    latex_code.append(r'\small\setlength\tabcolsep{2pt}')
    latex_code.append(r'\begin{longtable}{|p{2cm}|p{0.8cm}|p{5cm}|p{5cm}|p{9cm}|}')
    latex_code.append(r'\hline')
    latex_code.append(r'Model & Equation Number & Name & Latex Fragment & Input Variables \\')
    latex_code.append(r'\hline')
    latex_code.append(r'\endfirsthead')
    
    latex_code.append(r'\hline')
    latex_code.append(r'Model & Equation Number & Name & Latex Fragment & Input Variables \\')
    latex_code.append(r'\hline')
    latex_code.append(r'\endhead')

    current_equation_form = None

    # Remplir la table avec les données
    for index, row in df.iterrows():
        # Si l'Equation form change, insérer une ligne vide avec le nom de la forme centré
        if row['Equation form'] != current_equation_form:
            current_equation_form = row['Equation form']
            latex_code.append(r'\multicolumn{5}{c}{\textbf{' + str(current_equation_form) + r'}} \\')
            latex_code.append(r'\hline')
        
        model = row['Model']
        eq_number = row['Equation number']
        name = row['Name']
        latex_fragment = f"$${row['Latex fragment']}$$" if is_valid_latex(row['Latex fragment']) else "Invalid latex"  # Vérifier et inclure uniquement les fragments valides
        input_variable = row['Input variable']
        
        # Ajouter chaque ligne de données au tableau
        latex_code.append(f'{model} & {eq_number} & {name} & {latex_fragment} & {input_variable} \\\\')
        latex_code.append(r'\hline')
    
    # Fin de la table LaTeX
    latex_code.append(r'\end{longtable}')
    
    # Retourner le code LaTeX
    return '\n'.join(latex_code)


# Générer la table LaTeX
latex_table = generate_latex_table(df_filtered)

latex_table = latex_table.replace(',', ', ')
latex_table = latex_table.replace(' ,', ',')

# Écrire dans un fichier .tex
with open('../../../../Memoire/results/table_damage_functions.tex', 'w') as f:
    f.write(latex_table)

# Afficher le code LaTeX généré
print(latex_table)
```
