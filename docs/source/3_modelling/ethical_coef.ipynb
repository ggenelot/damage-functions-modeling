## Definition of the ethical coefficient

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

norm_constants = np.random.uniform(10000, 50000, 50)
revenue_per_cap = np.linspace(10000, 50000, 1000)
exponents = np.random.normal(0, 1, 50)
```

```python

# Create a color palette based on the exponent values
palette = sns.color_palette("seismic", len(norm_constants))


# Assuming norm_constant and revenue_per_cap are available as variables
for j, norm_constant in enumerate(norm_constants): 
    for i,  exponent in enumerate(exponents):
        # Calculate the ratio
        normalisation_value = (norm_constant / revenue_per_cap)**exponent

        # Plot the ratio
        sns.lineplot(x=revenue_per_cap, y=normalisation_value, estimator=None, alpha=0.01, color=palette[j])

#plt.plot(x=revenue_per_cap,  y=ratio)

sns.lineplot(x=revenue_per_cap, y=1, estimator=None, alpha=0.5, color='red')

plt.ylim(0, 6)
plt.xlabel('Revenue per capita')
plt.ylabel('Value')
plt.title('Norm Constant / Revenue per Cap')
plt.show()
```

```python
exponents = np.random.normal(0, 1, 100)


# Create a color palette going from -1 to 1
palette = sns.color_palette("PiYG", as_cmap=True)(np.linspace(-1, 1, len(exponents)))


# Assuming norm_constant and revenue_per_cap are available as variables

norm_constant = 25000
for i,  exponent in enumerate(exponents):
    # Calculate the ratio
    normalisation_value = (norm_constant / revenue_per_cap)**exponent

    # Plot the ratio
    # Find the closest color in the palette for the current exponent
    closest_color_index = np.argmin(np.abs(np.linspace(-1, 1, len(palette)) - exponent))
    sns.lineplot(x=revenue_per_cap, y=normalisation_value, estimator=None, alpha=0.3, color=palette[closest_color_index])

#plt.plot(x=revenue_per_cap,  y=ratio)

sns.lineplot(x=revenue_per_cap, y=1, estimator=None, alpha=0.9, color='black')

plt.ylim(0, 6)
plt.savefig('../../../../Memoire/results/exponent.png', dpi=300)

plt.show()
```
